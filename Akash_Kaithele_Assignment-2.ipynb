## 3. Perform the Below Visualizations. 
● Univariate Analysis 
● Bi - Variate Analysi s
● Multivariate Analysis

```python
# Univariate Analysis
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram for 'living area'
plt.figure(figsize=(10, 5))   # (width, height)
sns.histplot(dataset['living area'],color='blue')    # KDE=Kernel Density Estimation
plt.title('Histogram for Variable living area')
plt.show()

# Plot boxplot for 'number of bedrooms'
plt.figure(figsize=(10, 5))
sns.boxplot(x=dataset['number of bedrooms'], color='green')
plt.title('Boxplot for Variable number of bedrooms')
plt.show()
```

```python
# Scatter plot for 'living area' vs 'number of bedrooms'
plt.figure(figsize=(10, 5))
sns.scatterplot(x=dataset['living area'], y=dataset['number of bedrooms'], color='red')
plt.title('Scatter Plot for living area vs number of bedroom')
plt.show()

# Correlation heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(dataset[['living area','number of bedrooms','number of bathrooms']].corr())
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(dataset[['living area','number of bedrooms','number of bathrooms']])
plt.show()
```

##  5. Handle the Missing values.

```python
missing_values = dataset.isnull().sum()
print("Missing Values:")
print(missing_values)

#Drop rows with missing values
dataset_dropped = dataset.dropna()

#Fill missing values with a specific value like mean or median
dataset_filled = dataset.fillna(dataset.mean())  

#Interpolate missing values
dataset_interpolated = dataset.interpolate()

# Print the DataFrames after handling missing values
print("\nDataFrame after dropping rows with missing values:")
print(dataset_dropped)

print("\nDataFrame after filling missing values:")
print(dataset_filled)

print("\nDataFrame after interpolating missing values:")
print(dataset_interpolated)
```
